```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\Users\Asus\Drivering Analysis.xlsx'
data = pd.read_excel(file_path, engine='openpyxl')
data.head()
```

```python
data.info()
data.isnull().sum()
```

```python
data.describe()
```

```python
#Calculating total distance covered and fuel used across all trips.
total_distance = data['distance'].sum()
total_fuel_used = data['fuel_used'].sum()
print(f"Total Distance: {total_distance} km")
print(f"Total Fuel Used: {total_fuel_used} liters")
sns.scatterplot(data=data, x='distance', y='fuel_used', hue='speed', palette='viridis', s=100)
plt.show()
```

```python
#Calculating the average speed and average trip time.
avg_speed = data['speed'].mean()
avg_trip_time = data['trip_time'].mean()
print(f"Average Speed: {avg_speed} km/h")
print(f"Average Trip Time: {avg_trip_time} minutes")
```

```python
sns.histplot(data['speed'], kde=True, bins=10, color='red')
plt.show()
```

```python
sns.barplot(data=data,x='trip_id',y='speed',hue='trip_id',dodge=False,palette='inferno',errorbar=None)
sns.despine()
plt.xticks(rotation=90, fontsize=8)
plt.title("Average Speed by Trip", fontsize=14, fontweight='bold')
plt.xlabel("Trip ID", fontsize=12)
plt.ylabel("Speed (km/h)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=data,x='trip_id',y='fuel_used',hue='trip_id',palette='Set1',errorbar=None)
sns.despine()
plt.xticks(rotation=90, fontsize=8)
plt.title("Fuel Used Per Trip", fontsize=14, fontweight='bold')
plt.xlabel("Trip ID", fontsize=12)
plt.ylabel("Fuel Used (liters)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Selecting only numeric columns for the correlation matrix
numeric_data = data.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=14, fontweight='bold')
plt.show()
```

```python
sns.pairplot(data[['speed', 'distance', 'trip_time', 'fuel_used']], hue='speed', palette='Set1')
plt.show()
```
