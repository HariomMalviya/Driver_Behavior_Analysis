```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
data = pd.read_csv('../data/driver_data.csv')
print("Data Sample:")
print(data.head())

speed_limit = 70
high_acceleration = 3.0
high_braking = 0.3
sharp_turn = 3

# Analyze driver behavior
def analyze_driver_behavior(data):
    risky_drivers = []
    safe_drivers = []
    
    for driver_id in data['driver_id'].unique():
        driver_data = data[data['driver_id'] == driver_id]
        
        speeding_events = (driver_data['speed'] > speed_limit).sum()
        high_accel_events = (driver_data['acceleration'] > high_acceleration).sum()
        high_braking_events = (driver_data['braking'] > high_braking).sum()
        sharp_turn_events = (driver_data['cornering'] >= sharp_turn).sum()
        
        total_events = len(driver_data)
        risky_score = (speeding_events + high_accel_events + high_braking_events + sharp_turn_events) / total_events

        if risky_score > 0.3:
            risky_drivers.append((driver_id, risky_score))
        else:
            safe_drivers.append((driver_id, risky_score))
    
    return risky_drivers, safe_drivers

# Run analysis
risky_drivers, safe_drivers = analyze_driver_behavior(data)

# Display results
print("Risky Drivers:")
for driver_id, score in risky_drivers:
    print(f"Driver {driver_id} with Risk Score: {score:.2f}")

print("\nSafe Drivers:")
for driver_id, score in safe_drivers:
    print(f"Driver {driver_id} with Risk Score: {score:.2f}")

# --- Visualization ---
sns.set_theme(style="whitegrid")  # Apply a clean theme

# Convert results to DataFrame for easy plotting
risk_data = {
    "Driver ID": [driver_id for driver_id, _ in risky_drivers] + [driver_id for driver_id, _ in safe_drivers],
    "Risk Score": [score for _, score in risky_drivers] + [score for _, score in safe_drivers],
    "Status": ['Risky'] * len(risky_drivers) + ['Safe'] * len(safe_drivers)
}
df_risk = pd.DataFrame(risk_data)

# Plot: Risky vs Safe Drivers (Bar Chart)
plt.figure(figsize=(8, 5))
sns.countplot(x='Status', data=df_risk)
plt.title('Number of Risky vs Safe Drivers', fontsize=16)
plt.xlabel('Driver Status', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

# Plot: Distribution of Risk Scores (Histogram)
plt.figure(figsize=(8, 5))
sns.histplot(df_risk['Risk Score'], kde=True)
plt.title('Distribution of Driver Risk Scores', fontsize=16)
plt.xlabel('Risk Score', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

# Plot: Risk Scores by Driver Status (Boxplot)
plt.figure(figsize=(8, 5))
sns.boxplot(x='Status', y='Risk Score', data=df_risk)
plt.title('Risk Scores by Driver Status', fontsize=16)
plt.xlabel('Driver Status', fontsize=14)
plt.ylabel('Risk Score', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr_matrix = data[['speed', 'acceleration', 'braking', 'cornering']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Driving Behaviors')
plt.show()
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
plt.figure(figsize=(10, 6))
sns.lineplot(x='timestamp', y='speed', data=data, label='Speed')
plt.title('Driver Speed Over Time')
plt.xlabel('Time')
plt.ylabel('Speed')
plt.show()
```

```python
bins = [0, 0.1, 0.2, 0.3, 1]
labels = ['Low Risk', 'Moderate Risk', 'High Risk', 'Severe Risk']
df_risk['Risk Level'] = pd.cut(df_risk['Risk Score'], bins=bins, labels=labels)
sns.countplot(x='Risk Level', data=df_risk)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
df_risk['Cluster'] = kmeans.fit_predict(df_risk[['Risk Score']])
sns.scatterplot(x='Driver ID', y='Risk Score', hue='Cluster', data=df_risk)
plt.show()
```
